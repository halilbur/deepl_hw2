### Veri Kümesindeki Özelliklerin Anlamları ve Tipleri

1.  **school**: Öğrencinin okulu (binary: "GP" - Gabriel Pereira veya "MS" - Mousinho da Silveira) - *Kategorik (Binary)*
2.  **sex**: Öğrencinin cinsiyeti (binary: "F" - kadın veya "M" - erkek) - *Kategorik (Binary)*
3.  **age**: Öğrencinin yaşı (numeric: 15 ile 22 arası) - *Numerik (Kesikli)*
4.  **address**: Öğrencinin ev adresi tipi (binary: "U" - kentsel veya "R" - kırsal) - *Kategorik (Binary)*
5.  **famsize**: Aile büyüklüğü (binary: "LE3" - 3'ten küçük veya eşit veya "GT3" - 3'ten büyük) - *Kategorik (Binary)*
6.  **Pstatus**: Ebeveynlerin birlikte yaşama durumu (binary: "T" - birlikte yaşıyor veya "A" - ayrı) - *Kategorik (Binary)*
7.  **Medu**: Annenin eğitim durumu (numeric: 0 - yok, 1 - ilkokul (4. sınıf), 2 - 5-9. sınıf, 3 - lise veya 4 - yükseköğrenim) - *Numerik (Sıralı)*
8.  **Fedu**: Babanın eğitim durumu (numeric: 0 - yok, 1 - ilkokul (4. sınıf), 2 - 5-9. sınıf, 3 - lise veya 4 - yükseköğrenim) - *Numerik (Sıralı)*
9.  **Mjob**: Annenin mesleği (nominal: "teacher", "health" sağlık hizmetleri, "services" kamu hizmetleri (örn. idari veya polis), "at_home" ev hanımı veya "other" diğer) - *Kategorik (Nominal)*
10. **Fjob**: Babanın mesleği (nominal: "teacher", "health" sağlık hizmetleri, "services" kamu hizmetleri (örn. idari veya polis), "at_home" evde veya "other" diğer) - *Kategorik (Nominal)*
11. **reason**: Bu okulu seçme nedeni (nominal: "home" eve yakın, "reputation" okul itibarı, "course" ders tercihi veya "other" diğer) - *Kategorik (Nominal)*
12. **guardian**: Öğrencinin vasisi (nominal: "mother" anne, "father" baba veya "other" diğer) - *Kategorik (Nominal)*
13. **traveltime**: Evden okula seyahat süresi (numeric: 1 - <15 dk., 2 - 15-30 dk., 3 - 30 dk. - 1 saat, veya 4 - >1 saat) - *Numerik (Sıralı)*
14. **studytime**: Haftalık ders çalışma süresi (numeric: 1 - <2 saat, 2 - 2-5 saat, 3 - 5-10 saat, veya 4 - >10 saat) - *Numerik (Sıralı)*
15. **failures**: Geçmiş ders başarısızlık sayısı (numeric: n eğer 1<=n<3 ise, aksi halde 4) - *Numerik (Sıralı/Kesikli)*
16. **schoolsup**: Ek okul desteği (binary: evet veya hayır) - *Kategorik (Binary)*
17. **famsup**: Aileden eğitim desteği (binary: evet veya hayır) - *Kategorik (Binary)*
18. **paid**: Ders konusuyla ilgili (Matematik veya Portekizce) ek ücretli dersler (binary: evet veya hayır) - *Kategorik (Binary)*
19. **activities**: Ders dışı aktiviteler (binary: evet veya hayır) - *Kategorik (Binary)*
20. **nursery**: Anaokuluna gitti mi (binary: evet veya hayır) - *Kategorik (Binary)*
21. **higher**: Yükseköğrenim görmek istiyor mu (binary: evet veya hayır) - *Kategorik (Binary)*
22. **internet**: Evde internet erişimi (binary: evet veya hayır) - *Kategorik (Binary)*
23. **romantic**: Romantik bir ilişkisi var mı (binary: evet veya hayır) - *Kategorik (Binary)*
24. **famrel**: Aile ilişkilerinin kalitesi (numeric: 1 - çok kötü ile 5 - mükemmel arası) - *Numerik (Sıralı)*
25. **freetime**: Okul sonrası boş zaman (numeric: 1 - çok az ile 5 - çok yüksek arası) - *Numerik (Sıralı)*
26. **goout**: Arkadaşlarla dışarı çıkma (numeric: 1 - çok az ile 5 - çok yüksek arası) - *Numerik (Sıralı)*
27. **Dalc**: Hafta içi alkol tüketimi (numeric: 1 - çok az ile 5 - çok yüksek arası) - *Numerik (Sıralı)*
28. **Walc**: Hafta sonu alkol tüketimi (numeric: 1 - çok az ile 5 - çok yüksek arası) - *Numerik (Sıralı)*
29. **health**: Mevcut sağlık durumu (numeric: 1 - çok kötü ile 5 - çok iyi arası) - *Numerik (Sıralı)*
30. **absences**: Okula devamsızlık sayısı (numeric: 0 ile 93 arası) - *Numerik (Kesikli)*
31. **G1**: Birinci dönem notu (numeric: 0 ile 20 arası) - *Numerik (Sürekli/Kesikli)*
32. **G2**: İkinci dönem notu (numeric: 0 ile 20 arası) - *Numerik (Sürekli/Kesikli)*
33. **G3**: Final notu (numeric: 0 ile 20 arası, hedef değişken) - *Numerik (Sürekli/Kesikli)*

```python
def calculate_success(row):
    # Her üç dersin de geçer notu 10 ve üzerinde kabul edilir
    if row['G1'] >= 10 and row['G2'] >= 10 and row['G3'] >= 10:
        return 'Başarılı'
    else:
        return 'Başarısız'

# Başarı durumunu her öğrenci için hesapla
df['Success'] = df.apply(calculate_success, axis=1)

# Başarı durumu grafiği
sns.countplot(x='Success', data=df, hue='Success', palette='Set2', legend=False)
plt.title('Başarılı ve Başarısız Öğrenci Dağılımı (Tüm Dersler)')
plt.xlabel('Başarı Durumu')
plt.ylabel('Öğrenci Sayısı')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grupları filtreleme
only_schoolsup = df[(df['schoolsup'] == 'yes') & (df['paid'] == 'no')]
only_paid = df[(df['schoolsup'] == 'no') & (df['paid'] == 'yes')]
no_support = df[(df['schoolsup'] == 'no') & (df['paid'] == 'no')]

# G3 ortalamaları
avg_only_schoolsup = only_schoolsup['G3'].mean()
avg_only_paid = only_paid['G3'].mean()
avg_no_support = no_support['G3'].mean()

# Grafik verileri
labels = ['Sadece School Support', 'Sadece Paid Class', 'Hiçbiri']
values = [avg_only_schoolsup, avg_only_paid, avg_no_support]

# Bar Chart çizimi
plt.figure(figsize=(8,6))
plt.bar(labels, values, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('G3 Ortalama Notları - Destek Türüne Göre')
plt.ylabel('G3 Ortalama Notu')
plt.ylim(0, 20)

# Her barın üzerine değer yaz
for i, val in enumerate(values):
    plt.text(i, val + 0.4, f"{val:.2f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

#### Uç Değer Tespiti: IQR Yöntemi

```python
if 'df_processed' in locals() and not df_processed.empty:
    # Uç değerleri tespit edilecek numerik sütunları seçelim (örnek)
    numeric_cols_for_outliers = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                                 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 
                                 'absences', 'G1', 'G2', 'G3']
    # Sadece df_processed içinde var olan sütunları alalım
    numeric_cols_for_outliers = [col for col in numeric_cols_for_outliers if col in df_processed.columns and df_processed[col].dtype in ['int64', 'float64']]

    print("IQR Yöntemi ile Uç Değer Tespiti:")
    for col in numeric_cols_for_outliers:
        Q1 = df_processed[col].quantile(0.25)
        Q3 = df_processed[col].quantile(0.75)
        IQR_val = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR_val
        upper_bound = Q3 + 1.5 * IQR_val
        
        outliers = df_processed[(df_processed[col] < lower_bound) | (df_processed[col] > upper_bound)]
        if not outliers.empty:
            print(f"'{col}' sütununda {len(outliers)} adet uç değer bulundu (IQR Yöntemi).")
            # print(f"  Alt Sınır: {lower_bound:.2f}, Üst Sınır: {upper_bound:.2f}")
            # print(f"  Bulunan Uç Değerler (ilk 5 tanesi): {outliers[col].head().values}")
        # else:
            # print(f"'{col}' sütununda IQR yöntemi ile uç değer bulunamadı.")

    # Görselleştirme: Örnek olarak 'absences' sütunu için Box Plot
    if 'absences' in df_processed.columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df_processed['absences'])
        plt.title("'absences' Sütunu İçin Box Plot (Uç Değer Gösterimi)", fontsize=15)
        plt.xlabel('Devamsızlık Sayısı', fontsize=12)
        plt.show()
else:
    print("İşlenmiş veri seti (df_processed) bulunamadı.")
```

### Uç Değerlere Ne Yapıldığı, Hangi İşlem(ler)in Uygulandığı

Uç değerlerle başa çıkma yöntemleri, verinin doğasına ve analizin amacına göre değişir:

1.  **Tutma (Keeping):** Eğer uç değerler gerçek ve anlamlı verileri temsil ediyorsa (örneğin, bir dolandırıcılık tespit senaryosunda uç değerler önemli olabilir) veya kullanılan model uç değerlere dayanıklıysa (örn: bazı ağaç tabanlı modeller), uç değerler olduğu gibi bırakılabilir.

2.  **Silme (Removing):** Eğer uç değerlerin veri giriş hatası olduğu düşünülüyorsa veya analizi/modeli olumsuz etkileyecek kadar aşırıysa ve sayısı azsa, bu değerleri içeren satırlar silinebilir. Ancak bu, bilgi kaybına yol açabilir.

3.  **Değer Atama / Dönüştürme (Assigning/Transforming):**
    * **Sınırlandırma (Capping/Winsorizing):** Uç değerler, IQR yöntemindeki alt ve üst sınırlara veya belirli bir persentil değerine (örn. %1 ve %99) çekilir. Bu, aşırı değerlerin etkisini azaltır ama veriyi bir miktar değiştirir.
    * **Logaritmik Dönüşüm (Log Transformation):** Özellikle sağa çarpık (pozitif skewness) dağılımlarda kullanılır. Büyük değerleri küçülterek dağılımı normale yaklaştırabilir ve uç değerlerin etkisini azaltabilir.
    * **Diğer Dönüşümler:** Box-Cox dönüşümü gibi daha genel dönüşümler de kullanılabilir.

**Uygulanan İşlem:**

*Bu kısım, yukarıdaki analiz sonucuna ve seçilen stratejiye göre doldurulmalıdır.*

Örnek senaryo:
* "'absences' sütunundaki aşırı yüksek devamsızlık değerleri, model performansını olumsuz etkileyebileceği ve bu değerlerin sayısının az olduğu göz önüne alınarak, IQR yöntemine göre üst sınır olan X değeri ile sınırlandırılmıştır (capping).
* "'G1' sütunundaki birkaç uç değer, veri giriş hatası olmadığı ve öğrenci performansının doğal bir varyasyonu olabileceği düşünüldüğünden olduğu gibi bırakılmıştır."

Aşağıda, örnek olarak 'absences' sütunu için capping (sınırlandırma) işlemi gösterilmiştir. **Bu kod bloklarını kendi veri setinizdeki uç değer durumuna ve stratejinize göre uyarlamanız gerekmektedir.**

```python
if 'df_processed' in locals() and not df_processed.empty:
    # df_outlier_handled = df_processed.copy() # Orijinal df_processed'ı korumak için kopya üzerinde çalışalım
    df_outlier_handled = df_processed # Direkt df_processed üzerinde işlem yapalım, dikkatli olunmalı

    # Örnek: 'absences' sütunundaki uç değerleri IQR sınırları ile capping (sınırlandırma)
    if 'absences' in df_outlier_handled.columns:
        col_to_cap = 'absences'
        Q1 = df_outlier_handled[col_to_cap].quantile(0.25)
        Q3 = df_outlier_handled[col_to_cap].quantile(0.75)
        IQR_val = Q3 - Q1
        lower_cap = Q1 - 1.5 * IQR_val
        upper_cap = Q3 + 1.5 * IQR_val
        
        # Alt ve üst sınırdan taşan değerleri sınır değerleri ile değiştir
        # Negatif devamsızlık olmayacağı için lower_cap'i 0'dan küçükse 0 yapabiliriz.
        lower_cap = max(0, lower_cap)

        original_min = df_outlier_handled[col_to_cap].min()
        original_max = df_outlier_handled[col_to_cap].max()
        
        df_outlier_handled[col_to_cap] = np.where(df_outlier_handled[col_to_cap] > upper_cap, upper_cap,
                                                np.where(df_outlier_handled[col_to_cap] < lower_cap, lower_cap,
                                                         df_outlier_handled[col_to_cap]))
        
        print(f"'{col_to_cap}' sütunundaki uç değerler sınırlandırıldı (Capping).")
        print(f"  Önceki Min: {original_min}, Yeni Min (Alt Sınır): {lower_cap:.2f}, Güncel Min: {df_outlier_handled[col_to_cap].min():.2f}")
        print(f"  Önceki Max: {original_max}, Yeni Max (Üst Sınır): {upper_cap:.2f}, Güncel Max: {df_outlier_handled[col_to_cap].max():.2f}")

        # İşlem sonrası Box Plot
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df_outlier_handled['absences'])
        plt.title("'absences' Sütunu İçin Box Plot (Capping Sonrası)", fontsize=15)
        plt.xlabel('Devamsızlık Sayısı (Sınırlandırılmış)', fontsize=12)
        plt.show()
    else:
        print("'absences' sütunu bulunamadı.")

    # Diğer sütunlar için de benzer işlemler yapılabilir veya farklı stratejiler uygulanabilir.
    # Şimdilik, diğer sütunlarda bir işlem yapılmadığını varsayalım.
else:
    print("İşlenmiş veri seti (df_processed) bulunamadı.")
```

### Farklı Özellikler İçin Herhangi Bir Kodlama (Encoding) İşlemi

```python
if 'df_scaled' in locals() and not df_scaled.empty:
    df_encoded = df_scaled.copy() # Bir önceki adımdaki (ölçeklenmiş) df'i kopyalayalım
                                # Eğer ölçeklendirme yapılmadıysa df_outlier_handled.copy() kullanılabilir

    # Binary kategorik özellikleri (yes/no veya iki farklı string) 0/1'e dönüştürme
    # Bu student veri setindeki binary sütunlar zaten genelde bu formatta (örn: GP/MS, F/M, U/R)
    # Ancak 'yes'/'no' içerenler varsa map ile dönüştürülebilir.
    binary_cols_map = {
        'schoolsup': {'yes': 1, 'no': 0},
        'famsup': {'yes': 1, 'no': 0},
        'paid': {'yes': 1, 'no': 0},
        'activities': {'yes': 1, 'no': 0},
        'nursery': {'yes': 1, 'no': 0},
        'higher': {'yes': 1, 'no': 0},
        'internet': {'yes': 1, 'no': 0},
        'romantic': {'yes': 1, 'no': 0}
    }
    
    print("Binary özellikler için map ile 0/1 dönüşümü (varsa):")
    for col, mapping in binary_cols_map.items():
        if col in df_encoded.columns and df_encoded[col].dtype == 'object': # Sadece object tipindeyse ve map edilecekse
             # Sütundaki tüm değerlerin mapping anahtarlarında olup olmadığını kontrol et
            if all(item in mapping.keys() for item in df_encoded[col].unique() if pd.notna(item)):
                df_encoded[col] = df_encoded[col].map(mapping)
                print(f"  '{col}' sütunu 0/1'e dönüştürüldü.")
            else:
                print(f"  '{col}' sütununda mapping için beklenmedik değerler var, dönüşüm atlandı.")
        elif col in df_encoded.columns and df_encoded[col].dtype != 'object':
             print(f"  '{col}' sütunu zaten numerik, map işlemi atlandı.")


    # Diğer ikili (binary) ama farklı etiketlere sahip (örn: M/F, GP/MS) sütunlar için Label Encoding veya pd.get_dummies(drop_first=True)
    # pd.get_dummies(drop_first=True) binary özellikler için tek bir sütun (0/1) üretir.
    binary_cols_get_dummies = ['school', 'sex', 'address', 'famsize', 'Pstatus']
    print("\nDiğer binary özellikler için pd.get_dummies(drop_first=True) ile dönüşüm:")
    for col in binary_cols_get_dummies:
        if col in df_encoded.columns and df_encoded[col].dtype == 'object':
            # Eğer sütunda sadece 2 benzersiz değer varsa drop_first=True mantıklıdır.
            if df_encoded[col].nunique(dropna=True) == 2:
                df_encoded = pd.get_dummies(df_encoded, columns=[col], prefix=col, drop_first=True, dtype=int)
                print(f"  '{col}' sütunu için get_dummies(drop_first=True) uygulandı.")
            else:
                 print(f"  '{col}' sütunu ikiden fazla değere sahip veya object değil, get_dummies(drop_first=True) atlandı.")
        elif col in df_encoded.columns and df_encoded[col].dtype != 'object':
             print(f"  '{col}' sütunu zaten numerik, get_dummies atlandı.")

    # Nominal (çok kategorili) özellikler için One-Hot Encoding
    nominal_cols = ['Mjob', 'Fjob', 'reason', 'guardian']
    print("\nNominal özellikler için pd.get_dummies ile One-Hot Encoding:")
    for col in nominal_cols:
        if col in df_encoded.columns and df_encoded[col].dtype == 'object':
            df_encoded = pd.get_dummies(df_encoded, columns=[col], prefix=col, drop_first=False, dtype=int) # drop_first=False tüm kategoriler için sütun oluşturur
            print(f"  '{col}' sütunu için One-Hot Encoding uygulandı.")
        elif col in df_encoded.columns and df_encoded[col].dtype != 'object':
             print(f"  '{col}' sütunu zaten numerik, One-Hot Encoding atlandı.")

    print("\nKodlama sonrası veri setinin ilk birkaç satırı:")
    display(df_encoded.head())
    print("\nKodlama sonrası veri setinin boyutları:", df_encoded.shape)
    print("\nKodlama sonrası veri tipleri:")
    df_encoded.info()

else:
    print("Ölçeklenmiş veri seti (df_scaled) bulunamadı.")
```
